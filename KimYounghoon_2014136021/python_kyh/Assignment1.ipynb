## 06-1 내가 프로그램을 만들 수 있을까?

```python
def GuGu(n):
    L=[]
    L.append(n*1)
    L.append(n*2)
    L.append(n*3)
    L.append(n*4)
    L.append(n*5)
    L.append(n*6)
    L.append(n*7)
    L.append(n*8)
    L.append(n*9)
    print(L)
```

```python
GuGu(2)
```

```python
def GuGu1(n):
    L=[]
    for i in range(1,10):
        L.append(n*i)
    print(L)
```

```python
GuGu1(2)
```

```python
def GuGu2(n):
    L=[]
    i=1
    while i<10:
        L.append(n*i)
        i=i+1
    print(L)
```

```python
GuGu2(2)
```

## 06-2 3과 5의 배수 합하기

10 미만의 자연수에서 3과 5의 배수를 구하면 3, 5, 6, 9이다. 이들의 총합은 23이다.
1000 미만의 자연수에서 3의 배수와 5의 배수의 총합을 구하라.

- 초기값을 1로 두고 루프를 돌려 1씩 증가시키는 방법

```python
num=1
while num<1000:
    print(num, end=' ')
    num+=1
```

- range함수를 이용해서 1~999를 출력하는 방법

```python
for num in range(1,1000):
    print(num, end=' ')
```

```python
sum=0
for num in range(1,1000):
    if num%3==0 or num%5==0:
        sum+=num
print(sum)
```

```python
sum=0
for num in range(1,1000):
    if num%3==0:
        sum+=num
    if num%5==0:
        sum+=num
print(sum)
```

- 3과 5의 배수가 이중으로 더해질 수 있습니다.

```python
sum=0
for num in range(1,1000):
    if num%3==0:
        sum+=num
    elif num%5==0:
        sum+=num
print(sum)
```

## 06-3 게시판 페이징하기

A 씨는 게시판 프로그램을 작성하고 있다. 그런데 게시물의 총 건수와 한 페이지에 보여줄 게시물 수를 입력으로 주었을 때 총 페이지수를 출력하는 프로그램이 필요하다고 한다.

> 총 페이지수 = 총 건수 / 한 페이지당 보여줄 건수 + 1

```python
def getTotalPage(m,n):
    return m//n+1
print(getTotalPage(5, 10)) 
print(getTotalPage(15, 10)) 
print(getTotalPage(25, 10))
print(getTotalPage(30, 10))
```

- 나눗셈을 할 때 소수점을 제거하기 위해 /대신 //을 사용
- m과 n을 나눈 나머지가 0이면 1을 더하면 안됨

```python
def getTotalPage(m, n):
    if m % n == 0:
        return m // n
    else:
        return m // n + 1

print(getTotalPage(5, 10))
print(getTotalPage(15, 10))
print(getTotalPage(25, 10))
print(getTotalPage(30, 10))
```
