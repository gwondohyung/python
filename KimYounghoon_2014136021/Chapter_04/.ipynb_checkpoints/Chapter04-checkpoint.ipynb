```python
import pandas as pd
from glob import glob
```

```python
glob('C:/Users/icaru/chap04/excel/지역*.xls')
```

```python
stations_files = glob('C:/Users/icaru/chap04/excel/지역*.xls')
stations_files
```

```python
tmp_raw = []

for file_name in stations_files:
    tmp = pd.read_excel(file_name, header=2)
    tmp_raw.append(tmp)
    
station_raw = pd.concat(tmp_raw)
```

```python
station_raw.info()
```

```python
station_raw.head()
```

```python
stations = pd.DataFrame({'Oil_store':station_raw['상호'],
                        '주소':station_raw['주소'],
                        '가격':station_raw['휘발유'],
                        '셀프':station_raw['셀프여부'],
                        '상표':station_raw['상표'] })
stations.head()
```

```python
stations['구'] = [eachAddress.split()[1] for eachAddress in stations['주소']]
stations.head()
```

```python
stations['구'].unique()
```

```python
stations[stations['구'] == '서울특별시']
```

```python
stations.loc[stations['구'] == '서울특별시','구'] = '성동구'
stations['구'].unique()
```

```python
stations[stations['구'] == '특별시']
```

```python
stations.loc[stations['구'] == '특별시','구'] = '도봉구'
stations['구'].unique()
```

```python
stations[stations['가격'] == '-']
```

```python
stations = stations[stations['가격'] != '-']
stations.head()
```

```python
stations['가격'] = [float(value) for value in stations['가격']]
```

```python
stations.reset_index(inplace=True)
del stations['index']
```

```python
oil_price_top10 = stations.sort_values(by='가격', ascending=False).head(10)
oil_price_top10
```

```python
oil_price_bottom10 = stations.sort_values(by='가격', ascending=True).head(10)
oil_price_bottom10
```

```python
gmap_key = "AIzaSyALP3UWeCl4sFbx67GnE9sUiTFv79dOddg"
gmaps = googlemaps.Client(key=gmap_key)
```

```python
from tqdm import tqdm_notebook

lat=[]
lng=[]

for n in tqdm_notebook(oil_price_top10.index):
    try:
        tmp_add = str(oil_price_top10['주소'][n]).split('(')[0]
        tmp_map = gmaps.geocode(tmp_add)
        
        tmp_loc = tmp_map[0].get('geometry')
        lat.append(tmp_loc['location']['lat'])
        lng.append(tmp_loc['location']['lng'])
        
    except:
        lat.append(np.nan)
        lng.append(np.nan)
        print("Here is nan!")
        
oil_price_top10['lat'] = lat
oil_price_top10['lng'] = lng
oil_price_top10
```

```python
lat = []
lng = []

for n in tqdm_notebook(oil_price_bottom10.index):
    try:
        tmp_add = oil_price_bottom10['주소'][n].split('(')[0]
        tmp_map = gmaps.geocode(tmp_add)
        
        tmp_loc = tmp_map[0]['geometry']
        lat.append(tmp_loc['location']['lat'])
        lng.append(tmp_loc['location']['lng'])
    except:
        lat.append(np.nan)
        lng.append(np.nan)
        print("Here is nan!")
        
oil_price_bottom10['lat'] = lat
oil_price_bottom10['lng'] = lng
oil_price_bottom10
```

```python
ap = folium.Map(location=[37.5202, 126.975], zoom_start=10.5)

for n in oil_price_top10.index:
    if pd.notnull(oil_price_top10['lat'][n]):
        folium.CircleMarker([oil_price_top10['lat'][n], oil_price_top10['lng'][n]],
                           radius=15, color='#CD3181',
                           fill_color='#Cd3181').add_to(map)
    for n in oil_price_bottom10.index:
        if pd.notnull(oil_price_bottom10['lat'][n]):
            folium.CircleMarker([oil_price_bottom10['lat'][n],
                                 oil_price_bottom10['lng'][n]],
                                radius=15, color='#3186cc',
                                fill_color='#3186cc').add_to(map)
map
```
